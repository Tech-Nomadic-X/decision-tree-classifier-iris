# ðŸŒ³ Decision Tree Implementation using Scikit-learn

## ðŸ§  Objective
This project demonstrates how to implement a Decision Tree classifier using the Scikit-learn library on the Iris dataset. The goal is to classify flowers into three species based on their features and visualize the decision-making process.

Decision Trees are intuitive and powerful models that split the data based on feature thresholds, making them interpretable and effective for classification tasks.


ðŸŸ¦ 2. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, accuracy_score
```


ðŸŸ¦ 3. Load and Explore Dataset

```python
# Load iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Map numeric targets to class labels
df['species'] = df['species'].map({i: name for i, name in enumerate(iris.target_names)})

# Show dataset info
print("Dataset shape:", df.shape)
df.head()
```


ðŸŸ¦ 4. Visualize Feature Distributions

```python
# Pairplot to see feature distribution
sns.pairplot(df, hue='species')
plt.suptitle("Feature Distribution by Species", y=1.02)
plt.show()
```


ðŸŸ¦ 5. Prepare Data for Modeling

```python
# Features and labels
X = df.drop('species', axis=1)
y = df['species']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```


ðŸŸ¦ 6. Train the Decision Tree Model

```python
# Create and train the model
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
clf.fit(X_train, y_train)
```


ðŸŸ¦ 7. Visualize the Decision Tree

```python
# Visualize the decision tree
plt.figure(figsize=(15,10))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=clf.classes_)
plt.title("ðŸŒ³ Decision Tree Visualization", fontsize=16)
plt.show()
```


ðŸŸ¦ 8. Evaluate the Model

```python
# Predictions and performance
y_pred = clf.predict(X_test)

print("ðŸ” Classification Report:\n")
print(classification_report(y_test, y_pred))

print("âœ… Accuracy Score:", accuracy_score(y_test, y_pred))
```
